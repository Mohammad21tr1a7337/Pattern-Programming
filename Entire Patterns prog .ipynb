### Butterfly pattern

```python

n = 9
for i in range(1, n+1):
    
    for j in range(1, i+1):
        print("*", end="")
    
    for j in range(1, 2*(n-i)+1):
        print(" ", end="")
    
    for j in range(1, i+1):
        print("*", end="")
    print()


for i in range(n, 0, -1):
    
    for j in range(1, i+1):
        print("*", end="")
    
    for j in range(1, 2*(n-i)+1):
        print(" ", end="")
    
    for j in range(1, i+1):
        print("*", end="")
    print()
```

### Hour Glass

```python

n = 9 
for i in range(1, n+1):
    
    for j in range(1, i):
        print(" ", end="")
    
    for j in range(1, 2*(n-i)+1):
        print("*", end="")
    
    for j in range(1, i+1):
        print(" ", end="")
    print()


for i in range(n, 0, -1):
    
    for j in range(1, i+1):
        print(" ", end="")
    
    for j in range(1, 2*(n-i)):
        print("*", end="")
   
    for j in range(1, i):
        print(" ", end="")
    print()
```

## Diamond

```python
n = 9


for i in range(1, n+1):
    
    for j in range(n-i):
        print(" ", end="")
    
    for j in range(2*i-1):
        print("*", end="")
    print()


for i in range(n-1, 0, -1):
    
    for j in range(n-i):
        print(" ", end="")
    
    for j in range(2*i-1):
        print("*", end="")
    print()
```

```python
n = 9


for i in range(1, n+1):
    
    for j in range(n-i):
        print("*", end=" ")
    
    for j in range(2*i-1):
        print(" ", end=" ")
    for j in range(n-i):
        print("*", end=" ")   
    print()


for i in range(n-1, 0, -1):
    
    for j in range(n-i):
        print("*", end=" ")
    
    for j in range(2*i-1):
        print(" ", end=" ")
    for j in range(n-i):
        print("*", end=" ")
        
    print()
```

### Alphabets 

```python
a=9
for row in range(a):
    for col in range(a):
        if row==0 or row==a-1 or col==0 or col==a-1:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

```python
a=9
for row in range(a):
    for col in range(a):
        if row==0 or row==(a-1)/2 or col==0 or col==a-1:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

```python
a=9
for row in range(a):
    for col in range(a):
        if row==0 or row==(a-1)/2 or col==0 :
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```

```python
a = 9  

for row in range(a):
    for col in range(a):
        
        if row == 0 or row == a-1 or (row + col) == a-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

### Floyd's Triangle

```python
a=6
count=1
for i in range(a):
    for j in range(i):
        print(count,end=' ')
        count+=1
    print()
```

### Pascal Triangle

```python
n = 5 

for i in range(n):
    
    for j in range(n - i - 1):
        print(" ", end="")
    
    val = 1
    for j in range(i + 1):
        print(val, end=" ")
        val = val * (i - j) // (j + 1)
    print()
```
